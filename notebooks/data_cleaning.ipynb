## **Часть 3. Анализ и обработка пропусков**

### **3.1 director**

В столбце `director` аномалий нет, но пропущено около 30% значений, что составляет почти треть. Удалять данный столбец не стоит, потому что мы потеряем более 2.5 тыс. строк, а в этих записях может быть нужная информация для анализа по другим метрикам.

```python
df.groupby('type')['director'].apply(lambda x: x.isna().mean() * 100)
```

В ячейке выше мы посчитали процент пропусков по каждому типу. Выяснилось, что у фильмов всего 3% пропусков (данные почти полные, поэтому по ним можно проводить аналитику), а у ТВ-шоу - 91%.

И это неудивительно, потому что режиссеров обычно не указывают для ТВ-шоу. Поэтому все пропуски в столбце `director` мы заполним значением "Unknown", данный столбец удалять не нужно.

```python
df['director'] = df['director'].fillna('Unknown')
```

### **3.2 cast**

```python
df[df['cast'].isna()]['listed_in'].value_counts().head(15)
```

Мы видим, что пропуски по большей части в документальных фильмах/сериалах, в них актеров обычно не указывают.

Аномалий в `cast` нет, поэтому пустые значения заполним "Unknown".

```python
df['cast'] = df['cast'].fillna("Unknown")
```

### **3.3 country**

Пустые значения заполним "Unknown", а так же уберем лишние пробелы с запятыми в значениях `, South Korea` и `, France, Algeria`

```python
df['country'] = df['country'].str.strip(', ')
```

```python
df['country'] = df['country'].fillna("Unknown")
```

### **3.4 rating**

```python
list(df['rating'].unique())
```

Если мы выведем все уникальные значения рентинга, то увидим аномалию `['74 min', '84 min', '66 min']`.

Вставим значения "NR" вместо пустых ячеек, а также заменим на "NR" аномальные ячейки. 

```python
mask = df['rating'].str.contains('min', na=False)  # аналог WHERE rating LIKE '%min%' 
df.loc[mask, 'duration'] = df.loc[mask, 'rating']
df.loc[mask, 'rating'] = 'NR'
```

```python
df['rating'] = df['rating'].fillna("NR")
```

### **3.5 date_added**

Колонку `date_added` мы разобьем на три: `day_added`, `month_added`, `year_added`.

"Unknown" вставим вместо пустых ячеек.

```python
df['date_added'] = df['date_added'].str.strip() 
parsed = pd.to_datetime(df['date_added'], format='%B %d, %Y')  

df['day_added'] = parsed.dt.day.astype('Int64').astype(object)
df['month_added'] = parsed.dt.month.astype('Int64').astype(object)
df['year_added'] = parsed.dt.year.astype('Int64').astype(object)

mask_missing = df['date_added'].isna()
df.loc[mask_missing, ['day_added', 'month_added', 'year_added']] = 'Unknown'

df = df.drop(columns=['date_added'])
```

### **3.6 duration**

В данном случае колонку `duration` разобьем на две `duration_value` и `duration_unit`.

В пропуски заполним "Unknown".

```python
splited_duration = df['duration'].str.split(' ', expand=True)
df['duration_value'] = splited_duration[0]
df['duration_unit'] = splited_duration[1]

df = df.drop(columns=['duration'])
```
